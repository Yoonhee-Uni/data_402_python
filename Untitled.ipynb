```python
data = sns.load_dataset('titanic')
```

```python
df = pd.DataFrame(data)
df
```

```python
# checking missing values. missing values in deck but i will leave as it is for now
df.isnull()
```

```python
x = df.drop(columns = ['class', 'who', 'adult_male', 'embark_town', 'alive', 'alone'])
```

```python
print(x)
```

```python
x.drop_duplicates()
```

```python
x.rename(columns = {'parch' : 'parent/children'})
```

```python
# fill the Null age with average
average_age = x['age'].mean()
x['age'] = x['age'].fillna(average_age)
print(x)
```

```python
#basic descriptive and statistics for each column
x.describe()
```

```python
x.groupby(by="pclass")['age'].mean().round(2)
```

```python
x.groupby(by="pclass")['fare'].mean().round(2)
```

```python
# Add a new column showing the fare as a percentage of total
total_fare = x['fare'].sum().round(2)
print(total_fare)
x.assign(percentage_of_total_fare = (x['fare']/total_fare)*100 )
```
